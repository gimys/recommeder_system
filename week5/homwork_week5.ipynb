```python
import pandas as pd
import numpy as np
```

```python
pred = pd.read_csv('predictions.csv')
rating = pd.read_csv('ratings.csv')
```

```python
pred
```

```python
rating
```

```python
df_pred = pred.drop(columns='Unnamed: 0')
```

```python
df_rating = rating.drop(columns='Unnamed: 0')
```

```python
df_error = df_rating - df_pred
df_error
```

> ## 1. MAE
### 모든 값

```python
calc_mae = df_error.abs().sum().sum() / df_error.count().sum()
calc_mae
```

### 사용자

```python
calc_user_mae = df_error.abs().sum(axis=1) / df_error.count(axis=1)

user_mae = pd.DataFrame({'user':rating['Unnamed: 0'], 'MAE':calc_user_mae})
user_mae
```

### 아이템

```python
calc_item_mae = df_error.abs().sum() / df_error.count()

item_mae = pd.DataFrame({'MAE':calc_item_mae})
item_mae
```

> ## 2. MSE
### 모든 값

```python
mse = (df_error**2).sum().sum() / df_error.count().sum()
mse
```

### 사용자

```python
calc_user_mse = (df_error**2).sum(axis=1) / df_error.count(axis=1)

user_mes = pd.DataFrame({'user':rating['Unnamed: 0'], 'MSE':calc_user_mse})
user_mes
```

### 아이템

```python
calc_item_mse = (df_error**2).sum() / df_error.count()

item_mes = pd.DataFrame({'MSE':calc_item_mse})
item_mes
```

> ## 3. RMSE
### 모든 값

```python
rmse = np.sqrt(mse)
rmse
```

### 사용자

```python
cal_user_rmse = pd.DataFrame({'user':rating['Unnamed: 0'], 'RMSE':np.sqrt(user_mes['MSE'])})
cal_user_rmse
```

### 아이템

```python
cal_item_rmse = pd.DataFrame({'RMSE':np.sqrt(item_mes['MSE'])})
cal_item_rmse
```
