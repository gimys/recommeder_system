># 3. 평점 예측하기 ( $\frac{\sum^{5}_{n=1} r_n w_n}{\sum^{5}_{n=1}w_n}$)

```python
def calc_ex3(neighbor):
    tmp_list = np.zeros((5, 100))
    for i in range(len(neighbor.index)):
        user_id = neighbor.index[i]
        tmp_user_rating = users[users['Unnamed: 0'] == int(user_id)].drop(columns=['Unnamed: 0']).values
        tmp_list[i] = (tmp_user_rating*neighbor.values[i])[0]
    tmp_numerator = list(map(lambda x: np.nansum(x), np.transpose(tmp_list)))
    tmp_list_denominator = np.sum(neighbor.values)
    return tmp_numerator/tmp_list_denominator
```

```python
tmp_df = pd.DataFrame(columns=['title'])
tmp_df['title'] = items['Unnamed: 0']
tmp_df['3867'] = calc_ex3(neighbor_3867)
tmp_df['89'] = calc_ex3(neighbor_89)
tmp_df.head()
```

># 5. 정규화 추가 협업필터링 ($\bar{r_u} + \frac{\sum^{5}_{n=1} (r_n - \bar{r_n}) w_n}{\sum^{5}_{n=1}w_n}$)

```python
users[users['Unnamed: 0'] == 3867]
```

```python
def calc_ex5(neighbor, all_users):
    user_mean = np.nanmean(all_users[users.columns[1:]].values)
    tmp_list = np.zeros((5, 100))
    for i in range(len(neighbor.index)):
        user_id = neighbor.index[i]
        tmp_user_rating = users[users['Unnamed: 0'] == int(user_id)].drop(columns=['Unnamed: 0']).values
        tmp_list[i] = ((tmp_user_rating - np.nanmean(tmp_user_rating))*neighbor.values[i])[0]
    tmp_numerator = list(map(lambda x: np.nansum(x), np.transpose(tmp_list)))
    tmp_list_denominator = np.sum(neighbor.values)
    return user_mean + tmp_numerator/tmp_list_denominator
```

```python
tmp_df = pd.DataFrame(columns=['title'])
tmp_df['title'] = items['Unnamed: 0']
tmp_df['3867'] = calc_ex5(neighbor_3867, users[users['Unnamed: 0'] == 3867])
tmp_df['89'] = calc_ex5(neighbor_89, users[users['Unnamed: 0'] == 89])
tmp_df.head()
```
