```python
w_over_idf = item_for_idf.values*idf.values
w_over_idf
```

```python
item_for_idf = pd.DataFrame(data = w_over_idf, columns=keyword)
item_for_idf
```

```python
users_pro_idf = pd.DataFrame(index=[0,1], columns=items_col)
users_pro_idf['Unnamed: 0'] = ['User 1', 'User 2']
users_pro_idf
```
